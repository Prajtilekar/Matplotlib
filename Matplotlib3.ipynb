```python
import matplotlib.pyplot as plt
```

- Univariate plot
- it is used to display rangewise data
- it gives a count with respect to range
- Ex. BloodPressure 80-120 

```python
# generate a data for 100 students 
# data should contain 0-100 percentaage
import numpy as np
per = np.random.randint(0,100,100)
per
```

```python
plt.hist(per)
```

(array([10., 10.,  9.,  8., 11.,  8., 13.,  9.,  8., 14.]),
 array([ 3. , 12.2, 21.4, 30.6, 39.8, 49. , 58.2, 67.4, 76.6, 85.8, 95. ]),
 <BarContainer object of 10 artists>)

 Here, meaining of above data is 2nd array represent range, from 3 to 12.2 we have 10 records as per 1st array. Autimatically 10 objects/ranges are created. We also can assign range with the use of bin as below e.g.

```python
# lets supply our own range values
plt.hist(per,bins=[50,60,70,80,90,100])
```

```python
new_data = np.append(per,[200,300,450,500])
new_data
```

```python
plt.hist(new_data)
```

These data away from Actual population called Outliers.

```python
# lets supply our own start and end value,
#Here if we want to check data between 80 to 120 range is used.
plt.hist(per,range=(80,120))
```

```python
# range + bins combination
# lets supply our own start and end value
# Data between given value also provide range in between this values.
plt.hist(per,range=(80,120),bins=[80,90,100])
```

Here in above graph, 80 to 90, 10 values are present and 90 to 100, 4 values are present.

```python
#combine multiple hist plot
#Here, alpha is used for color fading : Syntax - alpha=.3
plt.hist(per,range=(80,120),bins=[80,85,90,95],alpha=.3)
plt.hist(per,alpha=.3)
```

#Pie Plot
- Univariate plot
- Used to display %wise distribustion of data
- Ex. %of each share holder

```python
person = ['ajay','rajendra','vanashri']
money = [120000,450000,180000]
```

```python
plt.pie(money,labels=person);
```

```python
#display % using autopct
plt.pie(money,labels=person,autopct='%.f%%');
```

```python
#display % using autopct
plt.pie(money,labels=person,autopct='%.2f%%');
```

```python
#change color
plt.pie(money,labels=person,autopct='%.f%%',colors=['y','c','m']);
```

```python
person # It starts as anticlockwise direction
```

```python
money
```

```python
#separate pie with Mr. Ajay
plt.pie(money,labels=person,autopct='%.f%%',colors=['y','c','m'],
       explode=[0.3,0,0]);
```

```python
fg, ax = plt.subplots()

ax.pie(money,
    autopct = '%1.1f%%',
    colors = ['y','c','m'],
    frame = True
    )

ax.legend(person, loc = 'upper right')
   
ax.axis('equal')
plt.show()
```
