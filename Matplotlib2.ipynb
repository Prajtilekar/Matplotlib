```python
import matplotlib.pyplot as plt
```

```python
overs = [1,2,3,4,5]
runs_MI = [2,18,24,30,35]
```

```python
runs_RCB = [0,10,23,26,33]
```

**Combine 2 plots**

```python
plt.plot(overs,runs_MI,'o--g',label='MI Score',markerfacecolor='orange')
plt.legend()
```

here, **The plot() function in pyplot module of matplotlib library  used to draw points (markers) in a diagram. By default, the plot() function draws a line from point to point.**

- First we enter X-axis data then Y-axis data.
- o -> Used for draw marker.
- ( -- ) This line used for line format.
- label, used for write name in graph, here we need to use legend function to display label.


```python
plt.plot(overs,runs_MI,'o--g',label='MI Score',markerfacecolor='black')

plt.plot(overs,runs_RCB,'-.*r',label='RCB Score')

plt.legend() # it makes label in plot visible
```

# Bar plot
- Bi variate/Multivariate
- It is used to plot Category vs count/frequence/population
- Used to compare in multiple categories
- Ex. Statewise Population

```python
state = ['MAH','KA','AP','PB','UP']
count = [230,120,200,80,67]
```

```python
plt.bar(state,count)
```

```python
plt.plot(count)
```

```python
# change color
plt.bar(state,count,color=['blue','cyan','magenta','yellow','red'])
```

```python
#to solve the x axis category label problem increase the figure size
plt.figure(figsize=(18,16))
plt.bar(['Maharashtra','Karnataka','Andhrapradesh','Punjab','Uttarpradesh'],
        count,color=['blue','cyan','magenta','yellow','red'])
```

```python
# 2nd solution is to use xticks
plt.bar(['Maharashtra','Karnataka','Andhrapradesh','Punjab','Uttarpradesh'],
        count,color=['blue','cyan','magenta','yellow','red'])
plt.xticks(rotation=45);
```

```python
#Combine 2 bar plots
import numpy as np
count1= ['MAH','KA','AP','PB','UP']
Girls = [230,120,200,190,167]
Boys = [130,100,100,180,107]
```

```python
#Stack Bar graph
#w=width=0.4
plt.bar(count1,Girls)
plt.bar(count1,Boys)
plt.show()
```

```python
#Combine 2 bar plots.
w=0.4
x = ['Mech','CSE','EEE','IT']
Boys=[100,80,70,30]
Girls=[30,100,50,80]
bar1=np.arange(len(x))  #Set location on X axis.
#bar2=[i+w for i in bar1]
plt.bar(bar1,Boys,w,label='Boys')
plt.bar(bar1+0.4,Girls,w,label='Girls') #0.4 used for 2nd bar location
plt.xticks(bar1+0.2,x) #0.2 used for bar label location
plt.xlabel('Course')
plt.ylabel('Student')
plt.title('Student vs Count')
plt.legend()
plt.show()
```

```python
#reduce bar height
plt.barh(x,Girls)
```

```python
state = ['MAH','KA','AP','PB','UP']
count = [230,120,200,80,67]
plt.barh(['Maharashtra','Karnataka','Andhrapradesh','Punjab','Uttarpradesh'],
        count,color=['green','cyan','magenta','yellow','red'])
```

```python
#reduce bar height
plt.barh(['Maharashtra','Karnataka','Andhrapradesh','Punjab','Uttarpradesh'],
         count,height=0.4,color=['blue','cyan','magenta','yellow','red'])
```
