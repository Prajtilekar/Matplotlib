```python
#univariate line plot
runs = [2,0,5,6,10]
```

```python
import numpy as np
runs = np.cumsum(runs)
runs
```

```python
#will use plot() method 
plt.plot(runs)
```

**X-axis : Input/independant variable/Predictor Variable
y-axis :  Output/Dependent variable, Class/Target Variable**

**When we don't give X-axis values then in matplotlib it automatically takes stepping 0.5 values**

```python
 # Scale the values present on X axis
 plt.plot(runs)
 plt.xticks(range(5),range(1,6))
#Here, 1st range method used for change scale from 0.5 to 1 and second range used for labling.
```

```python
#if we want to label it
plt.xticks(range(5),['Over1','Over2','Over3','Over4','Over5']);
plt.show()
```

```python
#Bi variate plot
overs = ['Over_1','Over_2','Over_3','Over_4','Over_5']
display(overs,runs)
```

```python
#plt.plot(x,y)
plt.plot(overs,runs)
```

```python
#color: black,white,red,gree,blue,cyan, magenta, yellow,orange
plt.plot(overs,runs,color='green')
```

```python
#change line style
plt.plot(overs,runs,color='#e11142',linestyle='--')
```

```python
#change line width
plt.plot(overs,runs,color='#e11142',linestyle='--',linewidth=4)
```

```python
#mark the inetersection points using marker
plt.plot(overs,runs,color='#e11142',linestyle='--',
        marker='o')
```

```python
# display only markers
plt.plot(overs,runs,color='#e11142',linestyle='',
        marker='o')
```

```python
# display grid lines
plt.plot(overs,runs,color='#e11142',linestyle='',
        marker='o')
plt.grid()
```
